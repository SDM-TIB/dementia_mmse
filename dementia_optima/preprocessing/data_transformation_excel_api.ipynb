```python
import pandas as pd
import numpy as np
import scipy.stats as spstats
import matplotlib.pyplot as plt
from feature_transformation_cls import FeatureTransform

from ast import literal_eval

pd.set_option('display.max_columns', None)  
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', -1)


%matplotlib inline
```

```python
folder_path = '../../../datalcdem/data/optima/dementia_18July/class_mild_mod_sev/'
patient_df = pd.read_csv(folder_path+'patients_new.csv')
print ('Patient Column size:', len(patient_df.columns))
print (patient_df.shape, len(patient_df.columns), patient_df.columns)
display(patient_df.head(10))

#extract episode data and calculate duration
patient_episode_date_df = patient_df[['patient_id'] + [col for col in patient_df.columns if 'EPISODE_DATE' in col]]
display(patient_episode_date_df.head(10))

def calculate_duration_btw_dates(df, col1, col2 ):
    return ((pd.to_datetime(df[col1]) - pd.to_datetime(df[col2]))/  np.timedelta64(1, 'D'))/365.25

patient_episode_date_duration_df = patient_episode_date_df[['patient_id']].copy()
for episode_date_col in patient_episode_date_df.columns[2:]:
    patient_episode_date_duration_df[episode_date_col+'-EPISODE_DATE(year)'] = calculate_duration_btw_dates(patient_episode_date_df, episode_date_col, 'EPISODE_DATE')

# calculate negative values
print ('Negative Values in the patient_episode_date_duration_df: ', sum(patient_episode_date_duration_df.values.flatten() < 0))
display(patient_episode_date_duration_df.head(10))


# saving file to csv file and drop EPISODE_DATE
patient_df.drop(columns=[col for col in patient_df.columns if 'EPISODE_DATE' in col], inplace=True)
display(patient_df.head(100))
patient_df.to_csv(folder_path+ 'patient_new_df.csv', index=False)
```

```python
featuretransform = FeatureTransform(folder_path + 'patient_new_df.csv')
patient_df = featuretransform.get_data_frame()

# define class
patient_df['epDateFinal-epDateInicial(year)'] = ((pd.to_datetime(patient_df['epDateFinal']) 
                                                  - pd.to_datetime(patient_df['epDateInicial']))/  np.timedelta64(1, 'D'))/365.25
featuretransform.set_data_frame(patient_df)


# Drop columns After considering class with Slow, Normal, Fast conditions
patient_df['patient_class'] = 9
def calculate_class(patient_df, mmseInicialDiscretized, mmseFinalDiscretized, time1, time2, label):
    #return np.where((patient_df['mmseInicialDiscretized']==mmseInicialDiscretized) & (patient_df['mmseFinalDiscretized']==mmseFinalDiscretized) 
    #                & (patient_df['epDateFinal-epDateInicial(year)']>=time1) & (patient_df['epDateFinal-epDateInicial(year)']<time2), label, 9)  # largest value for class+1 = 3
    patient_df.loc[(patient_df['mmseInicialDiscretized']==mmseInicialDiscretized) & (patient_df['mmseFinalDiscretized']==mmseFinalDiscretized) 
              & (patient_df['epDateFinal-epDateInicial(year)']>=time1) & (patient_df['epDateFinal-epDateInicial(year)']<time2), 'patient_class'] = label

# normal, mild, moderate, severe
calculate_class(patient_df, 'mild', 'moderate', 0, 2, 0)
calculate_class(patient_df, 'mild', 'moderate', 2, 4, 1)
calculate_class(patient_df, 'mild', 'moderate', 4, max(patient_df['epDateFinal-epDateInicial(year)'])+1, 2)
calculate_class(patient_df, 'moderate', 'severe', 0, 3, 3)
calculate_class(patient_df, 'moderate', 'severe', 3, 5, 4)
calculate_class(patient_df, 'moderate', 'severe', 5, max(patient_df['epDateFinal-epDateInicial(year)'])+1, 5)
calculate_class(patient_df, 'mild', 'severe', 0, 5, 6)
calculate_class(patient_df, 'mild', 'severe', 5, 9, 7)
calculate_class(patient_df, 'mild', 'severe', 9, max(patient_df['epDateFinal-epDateInicial(year)'])+1, 8)
# drop columns Which has class
patient_df = patient_df[patient_df['patient_class']!=9]


'''# Drop columns After considering class Mild, Moderate adn Severe
patient_df['patient_class']= patient_df['mmseInicialDiscretized']
patient_df = patient_df[patient_df['patient_class']!='normal']
patient_df.replace(['mild', 'moderate', 'severe'], [0,1,2], inplace=True)'''


display(patient_df.head(10))
display(patient_df.describe(include='all'))

patient_df = patient_df.drop(labels=['mmseInicialDiscretized', 'mmseFinalDiscretized', 'mmseInicial', 'mmseFinal', 'epDateInicial',
                                     'epDateFinal', 'epDateToSevereData', 'epDateToMildData', 'epDateToModerateData', 'epDateFinal-epDateInicial(year)' ], axis=1)
patient_df.info()


# set patient_df
featuretransform.set_data_frame(patient_df)

# Nominal Transfromation
# featuretransform.apply_feature_transform(column_name='mmseFinalDiscretized', func_type='oneHot', data_type_func='nominal')
[featuretransform.apply_feature_transform(column_name=col, func_type='oneHot', data_type_func='nominal') for col in columns_nominal]



# Bins Transformation
#featuretransform.apply_feature_transform(column_name='mmseFinal', func_type='oneHot', data_type_func='bin', n_bins=5, encode='ordinal', strategy='quantile')
[featuretransform.apply_feature_transform(column_name=col, func_type='oneHot', data_type_func='bin', n_bins=5, encode='ordinal', strategy='quantile') for col in columns_bins]

# List Transformation
# classses
disorderName_cui_cls = list(featuretransform.extract_value_frm_list(patient_df['disorderName_cui'].dropna().apply(literal_eval).tolist()))
drugName_cui_cls = list(featuretransform.extract_value_frm_list(patient_df['drugName_cui'].dropna().apply(literal_eval).tolist()))
classeslist = [disorderName_cui_cls, drugName_cui_cls]
print (disorderName_cui_cls)
print (drugName_cui_cls)
#featuretransform.apply_feature_transform(column_name='drugName_cui', func_type='multiLabel', classes=drugName_cui_cls, data_type_func='ordinal')
[featuretransform.apply_feature_transform(column_name=col, func_type='multiLabel', classes=clss, data_type_func='ordinal') for col, clss in zip(columns_list, classeslist)]
featuretransform.show_dataframe()

# set patient_df
patient_df = featuretransform.get_data_frame()
```
