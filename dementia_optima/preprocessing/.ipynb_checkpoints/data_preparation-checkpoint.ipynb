```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt


pd.set_option('display.max_columns', None)  
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', -1)


%matplotlib inline
```

```python
folder_path = '../../../data/dementia/dementia_01July/'

patient_df = pd.read_csv(folder_path + 'optima_patients.csv')
display(patient_df.head(5))
patient_df['Date_1'] = pd.to_datetime(patient_df.EPISODE_DATE)
patient_df = patient_df.sort_values(by='Date_1')
patient_df_merge_1 = patient_df.groupby('patient_id').agg(lambda x: x.tolist()).reset_index()
print(patient_df_merge_1.shape)
patient_df_merge_1.head(5)
```

```python
patient_df = pd.read_csv(folder_path + 'optima_patients_comorbidities.csv')
display(patient_df.head(10))

patient_df['Date_2'] = pd.to_datetime(patient_df.EPISODE_DATE)
patient_df = patient_df.sort_values(by='Date_2')
patient_df_merge_2 = patient_df.groupby('patient_id').agg(lambda x: x.tolist()).reset_index()
print(patient_df_merge_2.shape)
patient_df_merge_2.head(5)
```

```python
toxicitiespatient_df = pd.read_csv(folder_path + 'optima_patients_filtered.csv')
display(patient_df.head(10))

patient_df['Date_3'] = pd.to_datetime(patient_df.EPISODE_DATE)
patient_df = patient_df.sort_values(by='Date_3')
patient_df_merge_3 = patient_df.groupby('patient_id').agg(lambda x: x.tolist()).reset_index()
print(patient_df_merge_3.shape)
patient_df_merge_3.head(5)
```

```python
patient_df = pd.read_csv(folder_path + 'optima_patients_treatments.csv')
display(patient_df.head(10))

patient_df['Date_4'] = pd.to_datetime(patient_df.EPISODE_DATE)
patient_df = patient_df.sort_values(by='Date_4')
patient_df_merge_4 = patient_df.groupby('patient_id').agg(lambda x: x.tolist()).reset_index()
print(patient_df_merge_4.shape)
patient_df_merge_4.head(5)
```

```python
patient_df_merge_1 = patient_df_merge_1.merge(patient_df_merge_2,on='patient_id', how='outer')
display (patient_df_merge_1.shape)
patient_df_merge_1 = patient_df_merge_1.merge(patient_df_merge_3,on='patient_id', how='outer')
display (patient_df_merge_1.shape)
patient_df_merge_1 = patient_df_merge_1.merge(patient_df_merge_4,on='patient_id', how='outer')
display (patient_df_merge_1.shape)
patient_df_merge_1.head(5)
```

```python
patient_df_merge = patient_df_merge_1.drop(axis=1,labels=[c for c in patient_df_merge_1.columns if 'Date' in c or 'DATE' in c ])
print (patient_df_merge.shape)
display(patient_df_merge.head(5))
```
