```python
import pandas as pd
```

```python
df_treatments = pd.read_csv('../../../datalcdem/data/optima/dementia_18July/Treatments.csv')
```

```python
df_treatments.head(10)
```

```python
df_treatments_types = pd.read_csv('../../../datalcdem/data/optima/dementia_18July/defined_drugs_treatments.csv')
df_treatments_types['concept'] = df_treatments_types['concept'].apply(lambda x : x.capitalize())
df_treatments_types['group'] = df_treatments_types['group'].apply(lambda x : x.capitalize())
df_treatments_types.columns = ['name', 'CUI_ID', 'Group']
```

```python
df_treatments_types.head(10)
```

```python
df_dementia_treatments_types = df_treatments.merge(df_treatments_types, on=['CUI_ID'], how='left')
df_dementia_treatments_types.head(10)
```

```python
df_dementia_treatments_types.columns=['Name_x','CUI_ID', 'DBpediaURL', 'Name_y','Group']
```

```python
df_dementia_treatments_types = df_dementia_treatments_types[['Name_x','Name_y', 'CUI_ID', 'DBpediaURL','Group']]
```

```python
df_dementia_treatments_types.head(10)
```

```python
df_dementia_treatments_types.to_csv('../../../datalcdem/data/optima/dementia_18July/dementia_drugs_treatments_types.csv', index=False)
```
