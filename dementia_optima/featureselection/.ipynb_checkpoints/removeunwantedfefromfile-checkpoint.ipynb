```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as spstats
import sklearn.preprocessing
pd.set_option('display.max_columns', None)  
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', -1)

from sklearn.preprocessing import Binarizer
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction import FeatureHasher
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import KBinsDiscretizer
import pickle

%matplotlib inline
```

```python
class PatientData:
    def __init__(self, file, **kwargs):
        self.file = file
        self.df = pd.read_csv(file, sep=kwargs['sep'])
    
    
    def getFile(self, **kwargs):
        return self.file
    
    
    def plotData(self,plottype, **kwargs):
        pass
    
    
    def get_data_frame(self):
        return self.df
    
    
    def add_column_df(self, column_name, feature_arr, new_column_name='-Transformation'):
        #feature_arr_T = feature_arr.transpose()
        #[self.df.insert(loc=df.columns.get_loc(column_name)+1+i, column=column_name[0:2]+str(i), value=new_col) 
                                                #for i, new_col in zip(range(len(feature_arr_T)), feature_arr_T)]
        self.df.insert(loc=self.df.columns.get_loc(column_name)+1, column=column_name+new_column_name, value=feature_arr.tolist())
        #self.df = self.df.assign(e=feature_arr)
        
    
    def show_dataframe(self, n=5):
        display(self.df.head(n))
```

```python
#Discription of Data
folder_path = '../../../data/lungcancer/Transformation_survival_stage_cls_19Jun/'
pdata_disc = PatientData(folder_path + 'dataset_description.csv', sep='\t')
#wantedFeatureList = ['patient_id', 'surgery-True', 'surgery-False', 'chemoth_list-C0008838', 'status-0', 'toxicities-C0033774'] #correlation
wantedFeatureList = ['patient_id', 'stage-C0441766', 'stage-C0441767', 'stage-C0441771', 'chemoth-True', 'chemoth-False', 'stage-C0441772', 'surgery-False', 'surgery-True', 
                     'radiationth-True', 'radiationth-False', 'toxicities-C0004093', 'toxicities-C0042963', 'toxicities-C0032285', 'toxicities-C0027497', 
                     't_type-C0242379', 'status-0', 'chemoth_list-C0008838', 'chemoth_list-C0210657', 'toxicities-C0333355', 'chemoth_list-C0079083', 'tki_bool-False', 
                     'tki_bool-True']

pdf_disc = pdata_disc.get_data_frame() 
pdf_disc.shape
cols = pdf_disc.columns
arg_cols_list = []
tmp = []
for col in cols:
    tmp.append(col)
    if 'Transformed' in col:
        arg_cols_list.append(tmp)
        tmp=[]
for mdf in arg_cols_list:
    display(pdf_disc[mdf].head(1))
```

```python
# LCPatientData_Label
pdataLabel = PatientData(folder_path + 'dataset_description.csv', sep='\t')
#pdataLabel.show_dataframe()
display(pdataLabel.get_data_frame().describe(include='all'))

pdf_label = pdataLabel.get_data_frame()[['patient_id','Survival_Month', 'Survival_Class', 'Transformed Survival_Class']]
pdf_label.head(5)
```

```python
# Transformed LCPatientData
file_name = 'dataset_LCPatient'

pdata = PatientData(folder_path + file_name + '.csv', sep=',')
pdata.show_dataframe()
pdf = pdata.get_data_frame()


#Remove Unwanted Features
pdf = pdf[wantedFeatureList]

display(pdf.columns, pdf.shape)
display(pdf.describe())
```

```python
# Analysis data and remove unnecessary columns

#get column names


pdf_tran = pdf.transpose()
pdf = pdf_tran[pdf.max(axis=0) > 0.0].transpose()
display(pdf.head(5), pdf.shape)

pdf.to_csv(folder_path + file_name + '_fltrcols_unwanted' +'.csv', index=False)
```

```python
# survival month to category
def survivalMonthsCat(x):
    t = x//12
    if t < 4:
        return t
    return 4

pdf_label_copy = pdf_label.copy()

#quantile kbins for patients
from sklearn.preprocessing import KBinsDiscretizer
est = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='quantile')
pdf_label_copy['SurvivalMonthsCat'] = est.fit_transform(pdf_label[['Survival_Month']]) 
bin_edges = est.bin_edges_
print (bin_edges)

#category user defined 
#pdf_label_copy.loc[:,'survivalMonthsCat'] = pdf_label['survivalMonths'].apply(survivalMonthsCat)
#bin_edges = [0, 12, 24, 36, 48]

pdf_label_copy['SurvivalMonthsCatBin'] = pd.cut(pdf_label['Survival_Month'], np.concatenate(bin_edges, axis=0), right=False)

pdf_label = pdf_label_copy
pdf_label['SurvivalMonthsCat'].hist()
pdf_label['Transformed Survival_Class'].hist()

# merge dataframe based on id
pdf_surmon = pd.merge(pdf, pdf_label, on=['patient_id'] )
pdf_surmon_cols = pdf_surmon.columns.tolist()
pdf_data = pdf_surmon[pdf_surmon_cols[0:1] + pdf_surmon_cols[-5:] + pdf_surmon_cols[1:-5]]
display (pdf_data[pdf_data['patient_id']==1015121]['Survival_Month'], pdf_label[pdf_label['patient_id']==1015121]['Survival_Month'])
display (pdf_data.head(5), pdf_data.shape)

pdf_data.to_csv(folder_path + file_name + '_fltrcols_allfeatures_unwanted' +'.csv', index=False)
# patient data from dataframe
'''target_real = pdf_data['Survival_Month'].values
target = pdf_data['SurvivalMonthsCat'].values
target_names = pdf_data['SurvivalMonthsCatBin'].values
data = pdf_data.iloc[:,4:].values'''

target_real = pdf_data['Survival_Month'].values
target = pdf_data['Transformed Survival_Class'].values
target_names = pdf_data['Survival_Class'].values
data = pdf_data.iloc[:,6:].values
class_names = ['Long survival(stage III or IV and survival_months >=12)','Short survival(stage III or IV and survival_months <12)',
               'Long survival(stage I or II and survival_months >=24)', 'Short survival(stage I or II and survival_months <24)']
patient_data = {'data':data, 'target':target, 'target_names':target_names, 'target_real':target_real, 'class_names':class_names }
pickle.dump( patient_data, open(folder_path + 'patient_data_unwanted.pkl', "wb"))
```
